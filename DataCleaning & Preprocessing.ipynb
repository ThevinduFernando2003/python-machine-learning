Step 1 : Import necessary library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pypllot as plt
```

Step 2 : Read Dataset

```python
df = pd.read_csv("Life Expectancy Data.csv")
```

```python
df.head() #top 5 data
```

```python
df.tail() #bottom 5 data
```

Step 3 : Sanity check of data

```python
#Shape
df.shape
```

```python
#info
df.info()
```

```python
#finding missing value (null values count of each column)
miss_values = df.isnull().sum()
```

```python
 #percentage of missing values
percent_miss_values = miss_values / df.shape[0]*100 
```

```python
#findng duplicates
df.duplicated().sum()
```

```python
#identifying garbage values
for i in df.select_dtypes(include = "object").columns:
    print(df[i].value_counts())
    print("***"*10)
```

Step 4 : Exploratory data Analysis

```python
#descriptive analysis
df.describe().T
```

```python
df.describe(include="object")
```

```python
#histogram to understand the distribution
#import warnings
#warnings.filterwarnings("ignore")
for i in df.select_dtypes(include = "number").column:
    sns.histplot(data = df, x=i)
    plt.show()
```

```python
#Boxplot to identify outliers
import warnings
warnings.filterwarnings("ignore")
for i in df.select_dtypes(include = "number").column:
    sns.boxplot(data=df, x=i)
    plt.show()
```

```python
df.select_dtypes(include="number").columns
```

```python
#scatter plot to understand the relationship
for i in ['Year','Adult Mortality','Infant Deaths', 'Alcohol', 'Percentage expenditure',
          'Hepatitis B', 'Measles', 'BMI', 'Under-five Deaths', 'Polio', 'Total expenditure',
          'Diptheria', 'HIV/Aids', 'GDP', 'Population', 'Thinness 1-19 years', 'Thinness 5-9 years',
          'Income composition of Resources', 'Schooling']:
    sns.scatterplot(data=df, x=i, y='Life Expectancy')

#create a scatter plot for each combination
```

```python
#correlation with heatmap to intercept the relation and multicollinarity
df.select_dtypes(include="number").corr()
```
